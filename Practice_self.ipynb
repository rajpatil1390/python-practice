# **String indexing & slicing >**

- 0- s[:] --> print all string 
- 1- s[2:10] --> print from 2nd index to n-1 index which is 9th index 
- 2- s[ :10] --> s[0:10]-- print the value from start index to 9th index 
- 3- s[4: ] -->  print from 4th index till end or last 
- 4- s[1:10:3] --> print 3rd step index from 1st index to 9th inex -- forward index 
- 5- s[::1] - print only 1st step index from 1st to last  (step slicing  
- 6- s[::2] -- print only 2nd step index from 1st to last (step slicing 
- 7- s[::-1] -- backward step slicing 
- 8- s[::-2] -- -2 step backward slicing
